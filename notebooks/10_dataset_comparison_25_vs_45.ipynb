```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_25 = pd.read_csv("saudi_data_jobs_cleaned_25rows_initial.csv")

df_45 = pd.read_csv("saudi_data_jobs_cleaned.csv")
```

```python
print("25 Jobs Dataset:")
print(df_25.shape)

print("\n45 Jobs Dataset:")
print(df_45.shape)
```

```python
role_25 = df_25["role_category"].value_counts()

role_25
```

```python
role_45 = df_45["role_category"].value_counts()

role_45
```

```python
# Count role categories in both datasets

role_25 = df_25["role_category"].value_counts()

role_45 = df_45["role_category"].value_counts()


# Create comparison table

role_comparison = pd.concat(
    [
        role_25.rename("25 Jobs"),
        role_45.rename("45 Jobs")
    ],
    axis=1
)


# Replace missing values with 0 and convert to integer

role_comparison = role_comparison.fillna(0).astype(int)


# Display table

role_comparison
```

```python
# Calculate percentages for fair comparison

role_percentage = role_comparison.copy()


role_percentage["25 Jobs %"] = (
    role_percentage["25 Jobs"] / len(df_25) * 100
).round(1)


role_percentage["45 Jobs %"] = (
    role_percentage["45 Jobs"] / len(df_45) * 100
).round(1)


role_percentage
```

```python
role_comparison.plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Role Category Comparison: 25 vs 45 Jobs")
plt.xlabel("Role Category")
plt.ylabel("Number of Jobs")

plt.xticks(rotation=45)
plt.legend(title="Dataset")

plt.show()
```

```python
experience_25 = df_25["experience_category"].value_counts()

experience_45 = df_45["experience_category"].value_counts()


# Create comparison table

experience_comparison = pd.concat(
    [
        experience_25.rename("25 Jobs"),
        experience_45.rename("45 Jobs")
    ],
    axis=1
)


# Fix decimals

experience_comparison = (
    experience_comparison
    .fillna(0)
    .astype(int)
)


experience_comparison
```

```python
experience_percentage = experience_comparison.copy()


experience_percentage["25 Jobs %"] = (
    experience_percentage["25 Jobs"] / len(df_25) * 100
).round(1)


experience_percentage["45 Jobs %"] = (
    experience_percentage["45 Jobs"] / len(df_45) * 100
).round(1)


experience_percentage
```

```python
experience_comparison.plot(
    kind="bar",
    figsize=(10,5)
)


plt.title("Experience Category Comparison: 25 vs 45 Jobs")
plt.xlabel("Experience Category")
plt.ylabel("Number of Jobs")


plt.xticks(rotation=0)
plt.legend(title="Dataset")

plt.show()
```

```python
tools_25 = (
    df_25["tools"]
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)


tools_45 = (
    df_45["tools"]
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)
```

```python
tools_comparison = pd.concat(
    [
        tools_25.rename("25 Jobs"),
        tools_45.rename("45 Jobs")
    ],
    axis=1
)


tools_comparison = (
    tools_comparison
    .fillna(0)
    .astype(int)
)


tools_comparison
```

```python
tools_comparison.plot(
    kind="bar",
    figsize=(12,5)
)


plt.title("Top Tools Comparison: 25 vs 45 Jobs")
plt.xlabel("Tools")
plt.ylabel("Number of Mentions")

plt.xticks(rotation=45)

plt.show()
```

```python
skills_25 = (
    df_25["skills"]
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)


skills_45 = (
    df_45["skills"]
    .str.split(",")
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)
```

```python
skills_comparison = pd.concat(
    [
        skills_25.rename("25 Jobs"),
        skills_45.rename("45 Jobs")
    ],
    axis=1
)


skills_comparison = (
    skills_comparison
    .fillna(0)
    .astype(int)
)


skills_comparison
```

```python
skills_comparison.plot(
    kind="bar",
    figsize=(12,5)
)


plt.title("Top Skills Comparison: 25 vs 45 Jobs")
plt.xlabel("Skills")
plt.ylabel("Number of Mentions")

plt.xticks(rotation=45)

plt.show()
```

```python
role_comparison.to_csv("role_comparison_25_vs_45.csv")

experience_comparison.to_csv("experience_comparison_25_vs_45.csv")

tools_comparison.to_csv("tools_comparison_25_vs_45.csv")

skills_comparison.to_csv("skills_comparison_25_vs_45.csv")
```
