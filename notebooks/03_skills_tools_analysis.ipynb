```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Reemaalsharif/Saudi-Data-Internship-Readiness-Map/main/data/saudi_data_jobs_cleaned.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df[['skills', 'tools']].head()
```

```python
tools_series = df['tools'].str.split(',')
tools_exploded = tools_series.explode()
tools_cleaned = tools_exploded.str.strip()
tool_counts = tools_cleaned.value_counts()

tool_counts.head(10)
```

## Tools Analysis

This section analyzes the most frequently mentioned tools in Saudi data-related job postings.

```python
top_10_tools = tool_counts.head(10)

top_10_tools.plot(kind='bar', figsize=(10, 5))

plt.title('Top 10 Tools in Saudi Data Jobs')
plt.xlabel('Tool')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.show()
```

### Tools Insight

SQL is the most frequently mentioned tool in the dataset, followed by Python, Excel, and Power BI.

This suggests that students interested in data-related roles in Saudi Arabia should prioritize SQL and Python, while also building practical dashboarding skills using tools such as Power BI and Excel.

```python
skills_series = df['skills'].str.split(',')
skills_exploded = skills_series.explode()
skills_cleaned = skills_exploded.str.strip()
skill_counts = skills_cleaned.value_counts()

skill_counts.head(10)
```

## Skills Analysis

This section analyzes the most frequently mentioned skills in Saudi data-related job postings.

```python
top_10_skills = skill_counts.head(10)

top_10_skills.plot(kind='bar', figsize=(10, 5))

plt.title('Top 10 Skills in Saudi Data Jobs')
plt.xlabel('Skill')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.show()
```

### Skills Insight

The most frequently mentioned skills include trend analysis, data validation, data analysis, data governance, reporting, and problem solving.

This suggests that Saudi data-related roles require both technical analysis skills and business-focused skills such as reporting, validation, and problem solving.
