```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
FIGURES_DIR = Path("reports/figures")

FIGURES_DIR.mkdir(
    parents=True,
    exist_ok=True
)
```

```python
final_roadmap = pd.DataFrame({

    "Level": [

        "Beginner",
        "Beginner",
        "Beginner",
        "Intermediate",
        "Intermediate",
        "Intermediate",
        "Career Ready"

    ],

    "Focus_Area": [

        "Programming",
        "Database Skills",
        "Data Analysis",
        "Advanced Analytics",
        "Machine Learning",
        "Business Intelligence",
        "Portfolio and Job Preparation"

    ],

    "Recommended_Skills": [

        "Python Basics",
        "SQL and Database Fundamentals",
        "Excel, Pandas, Data Cleaning",
        "Advanced SQL, Statistics, Data Modeling",
        "Machine Learning Basics and Model Evaluation",
        "Power BI, Dashboard Design, Reporting",
        "Projects, GitHub, CV, Interview Preparation"

    ]

})


final_roadmap
```

```python
final_roadmap["Learning_Order"] = [

    1,
    2,
    3,
    4,
    5,
    6,
    7

]


final_roadmap
```

```python
final_roadmap.to_csv(
    "final_data_career_roadmap.csv",
    index=False
)
```

```python
plt.figure(figsize=(12,6))


plt.bar(
    final_roadmap["Focus_Area"],
    final_roadmap["Learning_Order"]
)


plt.xlabel(
    "Career Development Stage"
)


plt.ylabel(
    "Learning Order"
)


plt.title(
    "Final Data Career Roadmap for CS Students"
)


plt.xticks(
    rotation=45,
    ha="right"
)


plt.tight_layout()


plt.savefig(
    FIGURES_DIR / "final_data_career_roadmap.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
```
