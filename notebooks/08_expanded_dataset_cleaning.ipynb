```python
def clean_city(city):
    city = str(city).strip()

    if city in ["Riyadh", "الرياض"]:
        return "Riyadh"
    elif city in ["Jeddah", "جدة"]:
        return "Jeddah"
    elif city in ["Saudi Arabia", "السعودية"]:
        return "Not specified"
    elif city in ["King Abdullah Economic City"]:
        return "King Abdullah Economic City"
    elif city in ["Mahd Adh Dhahab"]:
        return "Mahd Adh Dhahab"
    else:
        return city

df["city_cleaned"] = df["city"].apply(clean_city)

df[["city", "city_cleaned"]].head(10)
```

```python
# Check cleaned city values
df["city_cleaned"].value_counts()
```

```python
# Check experience level values
df["experience_level"].value_counts()
```

```python
def categorize_experience(exp):
    exp = str(exp).lower().strip()

    if "internship" in exp or "co-op" in exp or "trainee" in exp:
        return "Internship"

    elif "fresh graduate" in exp:
        return "Entry Level"

    elif "entry" in exp:
        return "Entry Level"

    elif "0-2" in exp or "0 - 2" in exp:
        return "Entry Level"

    elif "1+" in exp or "2+" in exp or "2 years" in exp:
        return "Entry Level"

    elif "2-4" in exp or "2-5" in exp or "3+" in exp or "3-5" in exp or "3-6" in exp:
        return "Mid-Level"

    elif "4+" in exp or "4-6" in exp or "4-10" in exp or "minimum 4" in exp:
        return "Mid-Level"

    elif "minimum 5" in exp or "5 years" in exp or "senior" in exp:
        return "Senior"

    elif "associate" in exp or "assistant" in exp:
        return "Entry Level"

    elif "not specified" in exp or "not applicable" in exp or exp == "nan":
        return "Not Specified"

    else:
        return "Not Specified"

df["experience_category"] = df["experience_level"].apply(categorize_experience)

df[["experience_level", "experience_category"]].head(15)
```

```python
# Check experience category distribution
df["experience_category"].value_counts()
```

```python
# Check role types
df["role_type"].value_counts()
```

```python
def categorize_role(role):
    role = str(role).lower().strip()

    if "business intelligence" in role or "bi" in role or "power bi" in role:
        return "BI"

    elif "data scientist" in role or "data science" in role:
        return "Data Science"

    elif "ai" in role or "ml" in role or "machine learning" in role:
        return "AI"

    elif "data analyst" in role or "data analysis" in role:
        return "Data Analysis"

    elif "fraud" in role:
        return "Fraud Analytics"

    elif "governance" in role or "data management" in role or "data quality" in role or "stewardship" in role:
        return "Governance"

    elif "performance" in role or "operations" in role or "commercial" in role or "corporate performance" in role:
        return "Performance Analytics"

    elif "product" in role:
        return "Product Analytics"

    elif "marketing" in role:
        return "Marketing Analytics"

    elif "research" in role or "quantitative" in role or "market" in role:
        return "Research Analytics"

    elif "healthcare" in role or "medical" in role:
        return "Healthcare Analytics"

    elif "pricing" in role or "financial" in role or "control and reporting" in role:
        return "Business / Financial Analytics"

    elif "internship" in role or "trainee" in role or "co-op" in role:
        return "Internship / Training"

    else:
        return "Other"

df["role_category"] = df["role_type"].apply(categorize_role)

df[["role_type", "role_category"]].head(20)
```

```python
# Check role category distribution
df["role_category"].value_counts()
```
