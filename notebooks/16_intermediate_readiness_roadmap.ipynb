```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
FIGURES_DIR = Path("reports/figures")

FIGURES_DIR.mkdir(
    parents=True,
    exist_ok=True
)
```

```python
intermediate_roadmap = pd.DataFrame({

    "Stage": [

        "Advanced Data Analysis",
        "Advanced SQL and Databases",
        "Statistics and Modeling",
        "Machine Learning",
        "Advanced BI and Reporting",
        "Cloud and Data Platforms"

    ],

    "Skills": [

        "Advanced Pandas, Data Manipulation, Data Exploration",

        "Advanced SQL, Data Modeling, Database Design",

        "Statistics, Probability, Hypothesis Testing",

        "Machine Learning Basics, Model Evaluation",

        "Power BI Advanced Features, Dashboard Design",

        "Azure Data Services, Cloud Data Concepts"

    ],

    "Priority": [

        "High",
        "High",
        "Medium",
        "Medium",
        "High",
        "Medium"

    ]

})


intermediate_roadmap
```

```python
intermediate_roadmap["Learning_Order"] = [
    1,
    2,
    3,
    4,
    5,
    6
]


intermediate_roadmap
```

```python
intermediate_roadmap.to_csv(
    "intermediate_readiness_roadmap.csv",
    index=False
)
```

```python
plt.figure(figsize=(11,6))


plt.bar(
    intermediate_roadmap["Stage"],
    intermediate_roadmap["Learning_Order"]
)


plt.xlabel(
    "Learning Stage"
)


plt.ylabel(
    "Order"
)


plt.title(
    "Intermediate Data Career Roadmap"
)


plt.xticks(
    rotation=45,
    ha="right"
)


plt.tight_layout()


plt.savefig(
    FIGURES_DIR / "intermediate_readiness_roadmap.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
```

# Intermediate Roadmap Insights

The intermediate roadmap was created for learners who already have basic programming, SQL, and data analysis knowledge.

The roadmap focuses on developing stronger analytical skills, database knowledge, statistical understanding, machine learning foundations, advanced business intelligence, and cloud data concepts.

The goal is to help learners move from basic data analysis skills toward professional data roles.
