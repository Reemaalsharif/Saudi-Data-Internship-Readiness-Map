This notebook improves the visual presentation of the Saudi Data Jobs analysis using Seaborn and Matplotlib.

The goal is to create clearer charts for tools, skills, cities, role categories, and experience levels.

## Load Cleaned Dataset

This section loads the cleaned Saudi data jobs dataset from GitHub.

```python
url = "https://raw.githubusercontent.com/Reemaalsharif/Saudi-Data-Internship-Readiness-Map/main/data/saudi_data_jobs_cleaned.csv"

df = pd.read_csv(url)

df.head()
```

## Top 10 Tools

This chart shows the most frequently mentioned tools in the collected Saudi data-related job postings.

```python
tools_series = df['tools'].str.split(',')
tools_exploded = tools_series.explode()
tools_cleaned = tools_exploded.str.strip()
tool_counts = tools_cleaned.value_counts()

top_10_tools = tool_counts.head(10)
top_10_tools
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    x=top_10_tools.values,
    y=top_10_tools.index
)

plt.title('Top 10 Tools in Saudi Data Jobs')
plt.xlabel('Number of Mentions')
plt.ylabel('Tool')
plt.show()
```

### Tools Insight

SQL, Python, Excel, Power BI, and Tableau appeared frequently in the collected job postings.

This suggests that students interested in data-related roles should prioritize database skills, programming basics, and dashboarding tools.

```python
df = pd.read_csv("saudi_data_jobs_cleaned.csv")

print("Data loaded successfully")
df.head()
```

```python
df.info()
```

```python
from pathlib import Path

FIG_DIR = Path("reports/figures")
TABLE_DIR = Path("reports/tables")

FIG_DIR.mkdir(parents=True, exist_ok=True)
TABLE_DIR.mkdir(parents=True, exist_ok=True)

print("Folders are ready")
```

```python
df[["job_title", "tools"]].head(10)
```

```python
tools_long = df.copy()

tools_long["tools"] = tools_long["tools"].fillna("")

tools_long["tool"] = tools_long["tools"].str.split(r",|;|\|")

tools_long = tools_long.explode("tool")

tools_long["tool"] = tools_long["tool"].astype(str).str.strip()

tools_long = tools_long[tools_long["tool"] != ""].copy()

tools_long[["job_title", "tool"]].head(15)
```

```python
def normalize_tool(tool):
    tool = str(tool).strip()
    tool = " ".join(tool.split())
    key = tool.lower().replace("-", " ")

    mapping = {
        "powerbi": "Power BI",
        "power bi": "Power BI",
        "ms power bi": "Power BI",
        "microsoft power bi": "Power BI",

        "tableau": "Tableau",

        "excel": "Excel",
        "ms excel": "Excel",
        "microsoft excel": "Excel",

        "sql": "SQL",
        "mysql": "MySQL",

        "python": "Python",
        "r": "R",

        "pytorch": "PyTorch",

        "powerpoint": "PowerPoint",
        "power point": "PowerPoint",
        "ms powerpoint": "PowerPoint",

        "dashboard": "Dashboards",
        "dashboards": "Dashboards",

        "bi tools": "BI tools",
        "business intelligence tools": "BI tools"
    }

    return mapping.get(key, tool)

tools_long["tool_normalized"] = tools_long["tool"].apply(normalize_tool)

tools_long[["tool", "tool_normalized"]].drop_duplicates().head(30)
```

```python
tool_counts = tools_long["tool_normalized"].value_counts()

top_tools = tool_counts.reset_index()
top_tools.columns = ["tool", "count"]

top_tools.head(15)
```

```python
top_tools.to_csv(TABLE_DIR / "top_tools_counts.csv", index=False)

print("Saved:", TABLE_DIR / "top_tools_counts.csv")
```

```python
top_15_tools = top_tools.head(15)

plt.figure(figsize=(10, 6))

sns.barplot(
    data=top_15_tools,
    x="count",
    y="tool"
)

plt.title("Top Tools Mentioned in Saudi Data Job Postings")
plt.xlabel("Number of Mentions")
plt.ylabel("Tool")
plt.tight_layout()

plt.savefig(FIG_DIR / "top_tools_bar_chart.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
top_tools_line = top_tools.head(15).copy()

top_tools_line["rank"] = range(1, len(top_tools_line) + 1)

plt.figure(figsize=(10, 5))

sns.lineplot(
    data=top_tools_line,
    x="rank",
    y="count",
    marker="o"
)

plt.title("Tool Frequency Drop-off by Rank")
plt.xlabel("Tool Rank")
plt.ylabel("Number of Mentions")
plt.xticks(top_tools_line["rank"])
plt.tight_layout()

plt.savefig(FIG_DIR / "top_tools_line_chart.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
top_tools_line[["rank", "tool", "count"]]
```

```python
tools_by_role = pd.crosstab(
    tools_long["tool_normalized"],
    tools_long["role_category"]
)

top_10_tool_names = top_tools.head(10)["tool"].tolist()

tools_by_role = tools_by_role.loc[top_10_tool_names]

tools_by_role
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(
    tools_by_role,
    annot=True,
    fmt="d"
)

plt.title("Top Tools by Role Category")
plt.xlabel("Role Category")
plt.ylabel("Tool")
plt.tight_layout()

plt.savefig(FIG_DIR / "top_tools_by_role_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
tools_by_experience = pd.crosstab(
    tools_long["tool_normalized"],
    tools_long["experience_category"]
)

tools_by_experience = tools_by_experience.loc[top_10_tool_names]

tools_by_experience
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(
    tools_by_experience,
    annot=True,
    fmt="d"
)

plt.title("Top Tools by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Tool")
plt.tight_layout()

plt.savefig(FIG_DIR / "top_tools_by_experience_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
tools_by_role.to_csv(TABLE_DIR / "tools_by_role_category.csv")
tools_by_experience.to_csv(TABLE_DIR / "tools_by_experience_category.csv")

print("Analysis tables saved.")
```
