# Visualization Improvements

This notebook improves the visual presentation of the Saudi Data Jobs analysis using Seaborn and Matplotlib.

The goal is to create clearer charts for tools, skills, cities, role categories, and experience levels.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

## Load Cleaned Dataset

This section loads the cleaned Saudi data jobs dataset from GitHub.

```python
url = "https://raw.githubusercontent.com/Reemaalsharif/Saudi-Data-Internship-Readiness-Map/main/data/saudi_data_jobs_cleaned.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.shape
```

```python
df.columns
```

## Top 10 Tools

This chart shows the most frequently mentioned tools in the collected Saudi data-related job postings.

```python
tools_series = df['tools'].str.split(',')
tools_exploded = tools_series.explode()
tools_cleaned = tools_exploded.str.strip()
tool_counts = tools_cleaned.value_counts()

top_10_tools = tool_counts.head(10)
top_10_tools
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    x=top_10_tools.values,
    y=top_10_tools.index
)

plt.title('Top 10 Tools in Saudi Data Jobs')
plt.xlabel('Number of Mentions')
plt.ylabel('Tool')
plt.show()
```

### Tools Insight

SQL, Python, Excel, Power BI, and Tableau appeared frequently in the collected job postings.

This suggests that students interested in data-related roles should prioritize database skills, programming basics, and dashboarding tools.

## Top 10 Skills

This chart shows the most frequently mentioned skills in the collected Saudi data-related job postings.

```python
skills_series = df['skills'].str.split(',')
skills_exploded = skills_series.explode()
skills_cleaned = skills_exploded.str.strip()
skill_counts = skills_cleaned.value_counts()

top_10_skills = skill_counts.head(10)
top_10_skills
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    x=top_10_skills.values,
    y=top_10_skills.index
)

plt.title('Top 10 Skills in Saudi Data Jobs')
plt.xlabel('Number of Mentions')
plt.ylabel('Skill')
plt.show()
```

### Skills Insight

The most repeated skills include data analysis, reporting, dashboarding, validation, and problem solving.

This shows that data-related roles require both technical and business-focused skills.

## Jobs by City

This chart shows the distribution of job postings by city.

```python
city_counts = df['city_cleaned'].value_counts()

city_counts
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(
    x=city_counts.values,
    y=city_counts.index
)

plt.title('Saudi Data Jobs by City')
plt.xlabel('Number of Job Postings')
plt.ylabel('City')
plt.show()
```

### City Insight

Riyadh appears as the most common city in the dataset.

Some postings did not mention a specific city and were labeled as Not specified during the cleaning step.

## Jobs by Role Category

This chart shows how job postings are distributed across broader role categories.

```python
role_counts = df['role_category'].value_counts()

role_counts
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    x=role_counts.values,
    y=role_counts.index
)

plt.title('Job Postings by Role Category')
plt.xlabel('Number of Job Postings')
plt.ylabel('Role Category')
plt.show()
```

### Role Category Insight

The dataset includes different types of data-related roles, including Data Analysis, Data Science, BI, AI, Governance, Internship, and Other.

This supports the idea that data-related opportunities are broader than only Data Analyst or Data Scientist titles.

## Jobs by Experience Level

```python
plt.figure(figsize=(8, 5))

sns.barplot(
    x=experience_counts.values,
    y=experience_counts.index
)

plt.title('Job Postings by Experience Level')
plt.xlabel('Number of Job Postings')
plt.ylabel('Experience Level')
plt.show()
```
