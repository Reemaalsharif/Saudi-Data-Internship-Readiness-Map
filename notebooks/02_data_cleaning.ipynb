```python
import pandas as pd
```

```python
url = "https://github.com/Reemaalsharif/Saudi-Data-Internship-Readiness-Map/raw/refs/heads/main/data/saudi_data_jobs_dataset.csv"

df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
df['city'].value_counts()
```

```python
df['city_cleaned'] = df['city'].replace({
    'Saudi Arabia': 'Not specified'
})

df['city_cleaned'].value_counts()
```

```python
def categorize_experience(exp):
    exp = str(exp).lower()

    if 'intern' in exp or 'trainee' in exp:
        return 'Internship'
    elif 'entry' in exp or 'junior' in exp or 'fresh' in exp or 'assistant' in exp or 'analyst program' in exp:
        return 'Entry Level'
    elif 'mid' in exp or '2' in exp or '3' in exp or '4' in exp:
        return 'Mid-Level'
    elif 'senior' in exp or 'lead' in exp or 'manager' in exp or '5' in exp:
        return 'Senior'
    else:
        return 'Not Specified'

df['experience_category'] = df['experience_level'].apply(categorize_experience)

df['experience_category'].value_counts()
```

```python
def categorize_role(title):
    title = str(title).lower()

    if 'intern' in title or 'trainee' in title:
        return 'Internship'
    elif 'business intelligence' in title or 'bi' in title or 'power bi' in title:
        return 'BI'
    elif 'data scientist' in title or 'machine learning' in title:
        return 'Data Science'
    elif 'data analyst' in title or 'analysis' in title or 'analytics' in title:
        return 'Data Analysis'
    elif 'ai' in title or 'artificial intelligence' in title:
        return 'AI'
    elif 'governance' in title or 'data management' in title:
        return 'Governance'
    else:
        return 'Other'

df['role_category'] = df['job_title'].apply(categorize_role)

df['role_category'].value_counts()
```

```python
df[['job_title', 'city', 'city_cleaned', 'experience_level', 'experience_category', 'role_category']].head(10)
```

```python
df.to_csv('saudi_data_jobs_cleaned.csv', index=False)
```
