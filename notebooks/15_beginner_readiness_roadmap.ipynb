```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
FIGURES_DIR = Path("reports/figures")

FIGURES_DIR.mkdir(
    parents=True,
    exist_ok=True
)
```

```python
beginner_roadmap = pd.DataFrame({

    "Stage": [
        "Programming Foundation",
        "Database Foundation",
        "Data Analysis",
        "Data Visualization",
        "Business Intelligence"
    ],

    "Skills": [
        "Python Basics",
        "SQL",
        "Data Cleaning and Excel",
        "Pandas and Data Visualization",
        "Power BI and Dashboard Development"
    ],

    "Priority": [
        "High",
        "High",
        "High",
        "Medium",
        "High"
    ]

})


beginner_roadmap
```

```python
beginner_roadmap.to_csv(
    "beginner_readiness_roadmap.csv",
    index=False
)
```

```python
beginner_roadmap["Learning_Order"] = [
    1,
    2,
    3,
    4,
    5
]


beginner_roadmap
```

```python
plt.figure(figsize=(10,5))


plt.bar(
    beginner_roadmap["Stage"],
    beginner_roadmap["Learning_Order"]
)


plt.xlabel(
    "Learning Stage"
)


plt.ylabel(
    "Order"
)


plt.title(
    "Beginner Data Career Roadmap"
)


plt.xticks(
    rotation=45
)


plt.tight_layout()


plt.savefig(
    FIGURES_DIR / "beginner_readiness_roadmap.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
```

## Beginner Roadmap Insights

The beginner roadmap was created based on the findings from the Saudi Data Jobs analysis.

The roadmap prioritizes the foundational skills and tools most commonly required in data-related roles.

The recommended learning order starts with programming and database fundamentals, followed by analysis, visualization, and business intelligence skills.

This roadmap helps Computer Science students understand the practical steps required to prepare for entry-level data careers.
